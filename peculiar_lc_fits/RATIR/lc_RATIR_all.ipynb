```python
import numpy as np
import scipy
import matplotlib
import snpy
import os
import matplotlib.pyplot as plt
```

```python
for file in os.listdir():
    if file.endswith("EBV2.snpy"):
        s=snpy.get_sn(file)
        if "J" in s.restbands:
            print(s.get_max("J"))
        elif "Jspm_AB" in s.restbands:
            print(s.get_max("Jspm_AB"))
        elif "Jnot" in s.restbands:
            print(s.get_max("Jnot"))
        elif "HAWKIJ" in s.restbands:
            print(s.get_max("HAWKIJ"))
        elif "Jrc2" in s.restbands:
            print(s.get_max("Jrc2"))
        elif "J2m" in s.restbands:
            print(s.get_max("J2m"))
        elif "Jspm" in s.restbands:
            print(s.get_max("Jspm"))
        s.plot()
        plt.show()
```
